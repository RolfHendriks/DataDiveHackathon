## Check Correlations

```python
def show_correlations(data: pd.DataFrame):
    columns = data.columns
    #mask = np.triu_indices()
    sns.heatmap(data.corr(numeric_only = True), annot = True, square = True, cmap = 'coolwarm')

plt.figure(figsize = (10, 10), layout = 'constrained')
plt.subplot(2,1,1)
show_correlations(data_subnational)
plt.title('Correlations - Subnational')
plt.subplot(2,1,2)
show_correlations(data_multi)
plt.title('Correlations - Multidimensional')

```
